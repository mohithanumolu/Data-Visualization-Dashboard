```python
import pandas as pd
df = pd.read_csv("students.csv")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart: average marks per subject
avg_marks = df.groupby("Subject")["Marks"].mean()
avg_marks.plot(kind="bar", color="skyblue")
plt.title("Average Marks per Subject")
plt.ylabel("Marks")
plt.show()

# Line chart: marks trend
df.sort_values("Name", inplace=True)
plt.plot(df["Name"], df["Marks"], marker="o")
plt.title("Student Performance Trend")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()

# Heatmap: correlation
pivot = df.pivot_table(index="Name", columns="Subject", values="Marks")
sns.heatmap(pivot, annot=True, cmap="coolwarm")
plt.title("Marks Correlation Heatmap")
plt.show()
```
